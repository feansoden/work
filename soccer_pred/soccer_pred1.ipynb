```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
# import data
df = pd.read_csv("EPL_Soccer_MLR_LR.csv")
df.head()
#df.info()
#df.describe()
```

```python
df.drop(columns = ['Club','PlayerName']).corr()
```

```python
plt.scatter(df['Cost'], df['Score'])
```

```python
import numpy as np
X = np.array(df['Cost']).reshape(-1, 1) 
y = np.array(df['Score']).reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, test_size = 0.25)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LinearRegression 
regr = LinearRegression() 
  
regr.fit(x_train, y_train) 
print(regr.score(x_test, y_test))
```

```python
y_pred = regr.predict(x_test)
plt.scatter(x_test, y_test, color ='b') 
plt.plot(x_test, y_pred, color ='k') 
  
plt.show()
```

```python
print(y_pred)
```

```python
X = df[['DistanceCovered(InKms)', 'Goals',
       'ShotsPerGame', 'AgentCharges', 'BMI', 'Cost',
       'PreviousClubCost']]
y = df['Score']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.75,
                                                    test_size = 0.25, random_state = 100)
```

```python
x_train_with_intercept = sm.add_constant(x_train)
lr = sm.OLS(y_train, x_train_with_intercept).fit()
lr.summary()
```
